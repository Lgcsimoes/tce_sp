# Objective

This notebook objective is to scrape the spending and income data from São Paulo State Accounting Office (Tribunal de Contas do Estado de São Paulo) website. Data is separated in a zip file for each municipality and for each year.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

# Generates list of files to be downloaded

The script below generates text files with the URL list of each file to be downloaded, separated by year.

```python
for year in range(2008,2019):
    
    id_year = str(2019-year)
    text_file = open('arquivos_'+str(year)+'.txt', 'w')
    
    # Loads first webpage
    html = urlopen('http://transparencia.tce.sp.gov.br/municipios-csv?combine=&ano_exercicio='+id_year+'&page=0')
    bsObj = BeautifulSoup(html.read())
    print('Carregando dados do ano '+str(year)+':\n')
        
    # Extracts the maximum number of website pages for a given year
    last_page = bsObj.find_all('li', class_='pager__item pager__item--last')
    last_page = last_page[0].a['href']
    ii_max = int(last_page.split('=')[-1])
    print('Página 1/'+str(ii_max+1))
    
    # Loops through all expenses and income files from the first webpage
    items = bsObj.body.tbody.find_all('td', class_=['views-field views-field-nothing','views-field views-field-nothing-1'])
    for item in items:
        text_file.write(item.a['href'] + '\n')
        
    # Loops the remaining webpages for a given year
    for page in range(1, ii_max+1):
        print('Página '+str(page+1)+'/'+str(ii_max+1))
        html = urlopen('http://transparencia.tce.sp.gov.br/municipios-csv?combine=&ano_exercicio='+id_year+'&page='+str(page))
        bsObj = BeautifulSoup(html.read())

        # Loops through all expenses and income files from each webpage
        items = bsObj.body.tbody.find_all('td', class_=['views-field views-field-nothing','views-field views-field-nothing-1'])
        for item in items:
            text_file.write(item.a['href'] + '\n')
            
    # Closes the text file of a given year
    text_file.close()
```

# File download

In order to download all zip files from a given year, use the following command line on a UNIX shell:

`wget -N -P 2017 -i arquivos_2017.txt`
